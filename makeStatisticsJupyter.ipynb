```python
import pandas as pd
import numpy as np

alldata = pd.read_csv("/home/barbora/PycharmProjects/DP-SecuritySDN/xTS_5_xx.txt", delimiter=" ")
alldata.shape
```

```python
# data = alldata.loc[(alldata['Src_IP'] == '10.0.0.1') | (alldata['Src_IP'] == '10.0.0.2')]
data = alldata
data.head(20)
```

```python
data.shape
```

```python
new_pcktin = data.loc[data['State'] == 'NEW_PACKET_IN']
new_pcktin.head()
```

```python
new_pcktin.shape
```

```python
passed = data.loc[(data['State'] == 'PASSED') | (data['State'] == 'REJECTED') | (data['State'] == 'REJECTED_BLACKLISTED')]
passed.head()
```

```python
passed.shape
```

```python
import datetime, time
def time_difference(start, end):
#     start = '2016-08-15 ' + start
#     end = '2016-08-15 ' + end
    diff = datetime.datetime.strptime(end, "%H:%M:%S.%f") - datetime.datetime.strptime(start, "%H:%M:%S.%f")
    return float(diff.seconds * 1000000 + diff.microseconds)/1000000

new_passed = passed
merged = pd.DataFrame(columns=['Src_IP', 'Time', 'State'])
for i in new_pcktin.index:
    for j in new_passed.index:
        if new_pcktin['Src_IP'][i] == new_passed['Src_IP'][j]:
            passed_row = new_passed.loc[[j]]
            new_passed.drop(j, inplace=True)
            merged.loc[i] = [passed_row['Src_IP'][j], 
                             time_difference(new_pcktin['Time'][i], passed_row['Time'][j]),
                             passed_row['State'][j]]
            print 'Created row is', merged.loc[i]
            break
merged
```
